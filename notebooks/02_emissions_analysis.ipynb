# Emissions Analysis

"EMISSIONS BY COUNTRY", "Average emissions over years", "Average Emissions by Income Group".

```python
import sys
sys.path.append('..')

from src import config as cfg
from src.preprocessing import build_merged_data
from src import descriptive as desc

cfg.apply_theme_minimal()
merged_data = build_merged_data()
```

## Top 20 Countries with Highest Emissions in 2023

```python
top_emissions = desc.top_emissions_by_country(merged_data)
top_emissions.to_csv(cfg.TABLES_DIR / 'top20_emissions_2023.csv', index=False)
fig, ax = desc.plot_top_emissions_bar(top_emissions, save_path=cfg.FIGURES_DIR / 'top20_emissions_2023.png')
```

## Top 20 Countries with Highest Emissions Per Capita (2023)

```python
top_per_capita = desc.top_per_capita_emissions(merged_data)
top_per_capita.to_csv(cfg.TABLES_DIR / 'top20_per_capita_emissions_2023.csv', index=False)
fig, ax = desc.plot_top_per_capita_emissions(top_per_capita, save_path=cfg.FIGURES_DIR / 'top20_per_capita_emissions_2023.png')
```

## Average Emissions Over Selected Years (2000, 2010, 2020, 2023)

```python
avg_by_year = desc.average_emissions_selected_years(merged_data)
avg_by_year.to_csv(cfg.TABLES_DIR / 'average_emissions_selected_years.csv', index=False)
fig, ax = desc.plot_average_emissions_selected_years(avg_by_year, save_path=cfg.FIGURES_DIR / 'average_emissions_selected_years.png')
```

## Average Emissions by Income Group

```python
by_group = desc.average_emissions_by_income_group(merged_data)
by_group.to_csv(cfg.TABLES_DIR / 'average_emissions_by_income_group.csv', index=False)
fig, ax = desc.plot_average_emissions_by_income_group(by_group, save_path=cfg.FIGURES_DIR / 'average_emissions_by_income_group.png')
```
